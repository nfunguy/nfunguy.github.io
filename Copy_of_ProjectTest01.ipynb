````python
import os
from google.colab import userdata
from transformers import pipeline, logging

# 1. Silence unnecessary warnings to keep your console clean
logging.set_verbosity_error()

# 2. Authenticate using your Colab Secret
# Make sure you named your secret 'HF_TOKEN' in the sidebar!
try:
    hf_token = userdata.get('HF_TOKEN')
    print("✓ Hugging Face Token found.")
except:
    print("! Warning: HF_TOKEN not found in Secrets. Using guest access.")
    hf_token = None

# 3. Specify the model explicitly
# This prevents the "No model was supplied" warning
model_id = "distilbert-base-uncased-finetuned-sst-2-english"

print(f"Loading {model_id}...")
classifier = pipeline(
    "sentiment-analysis",
    model=model_id,
    token=hf_token
)

# 4. Run your AI logic
print("\n" + "="*40)
print("AI STARTUP PoC: SENTIMENT ANALYSIS")
print("="*40)

test_phrases = [
    "The GPU cloud is affordable and fast!",
    "I am struggling with the complex documentation.",
    "This AI startup is going to be huge."
]

results = classifier(test_phrases)

for text, result in zip(test_phrases, results):
    label = result['label']
    score = result['score']
    print(f"[{label}] ({score:.2%}) -> {text}")

print("="*40)
````

### What is happening here?

To help you understand the "AI plumbing" you're setting up, here is a visual representation of the process:

1. **The Tokenizer:** Since AI models only understand numbers, the first thing that happens is your text is broken into small chunks (tokens) and converted into a numerical ID.
2. **The Model:** The numbers are fed into the "Brain" (a Pre-trained Transformer) which processes the mathematical patterns.
3. **The Output:** The model spits out a probability (e.g., 99% sure this is POSITIVE), which the code turns back into a human-readable label for you.

That output you see is the "behind-the-scenes" work of the Hugging Face library. It’s essentially the AI downloading its "brain" and "dictionary" to your Colab machine.

Here is the breakdown of what those files actually are and why your AI startup pitch ended up sounding a bit... repetitive.

---

## 1. What are these files? (The "AI Brain" Components)

When you see those progress bars, Colab is downloading the pieces needed to make GPT-2 work. Think of it like assembling a new employee's desk:

| File Name | Real-World Comparison | What it actually does |
| --- | --- | --- |
| **`config.json`** | **The Blueprint** | Tells the computer how many "layers" the brain has (GPT-2 has 12). |
| **`model.safetensors`** | **The Memory/Weights** | This is the huge file (548MB) containing the 124 million "connections" the AI learned during training. **Safetensors** is a modern format that is faster and more secure than older versions. |
| **`vocab.json`** | **The Dictionary** | A list of every word or partial word the AI knows, mapped to a number ID. |
| **`merges.txt`** | **Grammar Rules** | Tells the AI how to combine small character chunks (like "learn" + "ing") into a single idea. |

---

## 2. Why is the output so repetitive?

In your result, the AI started repeating: *"We learn to learn, learning, learning, learning."* This is a classic issue called **Neural Text Degeneration**. It happens for two reasons:

1. **Model Size:** GPT-2 (Small) is a "lightweight" model. It’s like a smart toddler—it starts a sentence well but can easily get stuck in a "thinking loop" where it thinks the most likely next word is the one it just said.
2. **Greedy Decoding:** By default, the AI picks the #1 most probable next word every time. If it says "learning," the probability of saying "learning" again often goes *up*, creating an infinite loop.

---

## 3. The Underlying Python: Lists and Indexing

To get that text onto your screen, you used this Python logic:

```python
print(outputs[0]['generated_text'])

```

### The "Box within a Box" Structure:

1. **`outputs`**: This is a **List** (a box). Even if there is only one result, Python puts it in a list so it *could* hold more if you wanted.
2. **`[0]`**: This is the **Index**. In Python, counting starts at `0`. This tells the code: "Open the first box in the list."
3. **`['generated_text']`**: This is a **Dictionary Key**. Inside that first box is a label called `generated_text`. This tells Python: "Give me the actual words stored under this label."

---

## 4. How to Fix the "Looping"

If you want your startup pitch to be more "creative" and less "loopy," we can add two simple parameters to your Python code: `temperature` and `do_sample`.

**Try changing your generation line to this:**

```python
outputs = generator(
    prompt,
    max_length=100,
    do_sample=True,   # Tells the AI it's okay to take risks
    temperature=0.7   # Higher = more creative/random, Lower = more predictable
)

```
